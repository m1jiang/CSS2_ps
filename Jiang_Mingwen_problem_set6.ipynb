# 1.Load the data

```python
import pandas as pd

from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

X = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)

y = diabetes.target
```

# 2.Import other libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.linalg import inv
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# 3.Create the function


```python
def linear(X,y):
  #take x and y as in the input
  Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state = 1)
  #split the X and y into trin and tes using train_test_split from sklearn

  model = LinearRegression(fit_intercept = True)
  model.fit(Xtrain,ytrain)
#fit a linear regression model to the training X and traing y data
  y_model = model.predict(Xtest)
  #get the predicted y data for X test from the model
  

  #output
  print(str(X.columns[0]))
  print('Training R^2',model.score(Xtrain,ytrain))
  print('Testing R^2',model.score(Xtest,ytest))
  print('y = ',model.coef_[0],'*x + ',model.intercept_)


  #plotting the two sub plots
  fig,axes = plt.subplots(nrows = 1,ncols = 2)
  sns.scatterplot(x=Xtrain[Xtrain.columns[0]],y=ytrain,ax = axes[0])
  sns.lineplot(x = Xtrain[Xtrain.columns[0]],y = ytrain,ax = axes[0])

  sns.scatterplot(x = Xtest[Xtest.columns[0]],y = ytest,ax=axes[1])
  sns.lineplot(x = Xtest[Xtest.columns[0]],y = y_model,ax=axes[1])


  return
```

# 4.Try it out with some of the data

```python
linear(X[['age']],y)
```

# 5. Using for loops. Find the best single feature (i.e., column) in X that best predicts y. Make subplots

```python
#creating a list that contains all of the testing R^2

for i in range(X.shape[1]):
  linear(X[[X.columns[i]]],y)
  
print('The best predictor is s5.')
```
